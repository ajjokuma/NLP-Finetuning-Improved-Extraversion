```python
import json
import re

def remove_first_two_sentences(text):
    # Define a regex pattern to match sentences ending with . or ?
    sentence_pattern = r'[^.!?]*[.!?]'

    # Find all matches in the text
    sentences = re.findall(sentence_pattern, text)

    # Remove the first two sentences
    remaining_sentences = sentences[2:]

    # Join the remaining sentences back into a single string
    result = ''.join(remaining_sentences).lstrip()

    return result


with open('prompt-common_responses.json', 'r') as f:
    data = json.load(f)

output_data = [remove_first_two_sentences(item['output']) for item in data]

with open('please-base-common_responses.json', 'w') as f:
    json.dump(output_data, f, indent=4)
```
